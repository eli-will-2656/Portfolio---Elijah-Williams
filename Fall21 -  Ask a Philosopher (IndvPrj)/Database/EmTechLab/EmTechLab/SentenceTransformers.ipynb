In this notebook we will be using ðŸ¤—Huggingface's Sentence Tranformers to create word embeddings for our list of quotes

Unlike most most sentenceTransformers (like GenSim's Doc2Vec), this model is bidirectional (takes context from both sides during training). This is because Huggingface Sentence transformers are based on a pretrained [sentence-BERT](https://arxiv.org/pdf/1908.10084.pdf) network.

```python
%%capture
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2') #This is the fastest sentence transformer

#Our sentences we like to encode
sentences = ['This framework generates embeddings for each input sentence',
    'Sentences are passed as a list of string.',
    'The quick brown fox jumps over the lazy dog.']

#Sentences are encoded by calling model.encode()
embeddings = model.encode(sentences)

#Print the embeddings
for sentence, embedding in zip(sentences, embeddings):
    print("Sentence:", sentence)
    print("Embedding:", embedding)
    print("")
```
