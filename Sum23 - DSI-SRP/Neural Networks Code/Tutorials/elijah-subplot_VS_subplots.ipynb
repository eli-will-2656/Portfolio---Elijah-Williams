# Subplot vs Subplots
*What's the difference? How can I tell them apart?*

## matplotlib.pyplot.subplots()
Subplots is an object-oriented approach to creating multiple plots on a single figure. Subplots is a function that create a figure, and returns that figure and an array of axes. By default, only one `axes` object is added to figure, and returned by the subplots function

Function call: `plt.subplots((shape tuple), sharex, sharey, etc.)`

Returns: `fig, axs`


```python
# Creating simple plot
import matplotlib.pyplot as plt
import numpy as np

fig, ax1 = plt.subplots()

x = np.linspace(-1,1)
y = np.absolute(x)
z = np.sin(x)

# We plot things on axes, not figures
ax1.plot(x, y)
plt.show()
```

Note that we plot graphs on **`axes`** as opposed to the `figure` object itself

```python
# We can use subplots to create a figure with multiple axes

fig, (ax1, ax2) = plt.subplots(1,2, sharey=True, figsize=(4,2))
ax1.plot(x, y, 'r')
ax2.plot(x, z, 'b')
plt.show()
```

## matplotlib.pyplot.subplot()

Function call: `plt.subplot(nrow, ncol, index)`

Returns: `axs` created on the figure

This function is used to add an axes object at specified index in a figure, partitioned into a given number of rows and columns.
The index starts at 1 at the top left and increase from left right.

```python
# Creating a figure with 4 axes objects

ax1 = plt.subplot(2,2,1)
ax2 = plt.subplot(2,2,2)
ax3 = plt.subplot(2,2,3)

ax1.plot(x,y)
ax2.plot(x,x)
ax3.plot(z,y)

plt.show()
```

# Comparison

**For making multiple subplots of different types**

The `subplots` method create a figure and an array of axes, returning that figure and array. As such, it is a more object-oriented approach to creating multiple subplots.

**For use in iterations**

The `subplot` method create and returns a axes at a specific location in figure, and is much better suited for iterations, as you can directly index the axes object of the figure with multiplem subplots
