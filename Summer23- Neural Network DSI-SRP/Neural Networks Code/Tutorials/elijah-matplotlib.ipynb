**Axes** is an artist belonging to a figure, that contains a region for plotting data, and includes `Axis` objects to provide the black lines with ticks to scale the data. 

```python
# Every Axes object, you can change the title, ylabel, and xlabel
fig, ax = plt.subplots()

ax.plot(["Ronaldo","Messi","Neymar"], [1325, 954, 1234])
ax.set_xlabel("player")
ax.set_ylabel("goals")
ax.set_ybound(0, 1500)
ax.set_title("Top Soccer Goal Scorers")
```

**Axis** - object that set the scale and limits and generate ticks and axis labels
**Artist** - everything visible on a figure belongs to `Artist` class, and is drawn to the canvas. Most `artists` are tied to an Axes object, and thus cannot be share by multiple axes object.

Two ways to use Matplotlib:
* OOP - Create Figure and Axes objects and call methods on them to plot
* Application - Use pyplot module functions to automatically create figures and axes

```python
# OO-style

x = np.linspace(0,2,100) # linspace does include stop interval and use number of samples

fig,ax = plt.subplots(figsize=(5,2.7), layout='constrained')
ax.plot(x,x,label='linear')
ax.plot(x,x**2,label="quadratic")
ax.plot(x,x**3,label="cubic")
ax.set_label("x-axis")
ax.set_label("y-axis")
ax.set_title("linear, quadratic, cubic functions")
ax.legend(); # adds a legend
```

```python
# Equivalently by calling pyplot methods

x = np.arange(-4,4,.01) # arrange does not include stop value, and uses increments

plt.figure(figsize=(4, 2)) #figsize gives width and heigh  of figure
plt.plot(x,x,label="linear")
plt.plot(x,x**2,label="quadratic")
plt.xlabel("x values")
plt.ylabel("awesomeness")
plt.title("Elijah's awesomeness over time")
plt.legend();
```

## Making Helper Functions

```python
def my_plotter(ax, x, y):
    """
    A helper function to make a graph
    """
    out = ax.plot(x,y)
    return out

x = np.random.randn(100000)
y = np.random.randn(100000)

fig, ax = plt.subplots()
my_plotter(ax, x, y)
```

For every object displayed on the figure (`Artists`) you can modify their attributes when plotted either by 
1. using `plot` parameters
