# Visualization with matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as R

m = 0.0  # make sure these are floats
s = 1.0
x = np.arange(-4.0, 4.0, .01)  # returns an array of numbers, not including the stop value

x[799] # notice that this won't return a precise value

p = (1/np.sqrt(2*np.pi*(s**2))) * np.exp(-((x-m)**2)/2*(s**2)) # Gaussian distribution (normal curve)

plt.plot(x, p) # plotting two nparray on a plane
plt.xlabel('x')
plt.ylabel('p(x)')
plt.title(f"Normal Distribution (m={m}, s={s})"); # use semi-colon to suppress unnecessary output

# Saving plot to file
plt.savefig("plots\lnplot.pdf", dpi=300, format="pdf") # dpi = dots per inches (default value='figure')
plt.savefig("plots\lnplot.jpg", dpi=300, format="jpg")
plt.savefig('plots\lnplot200.tif', dpi=200, format='tif') # tif = tage image file
plt.savefig('plots\lnplot600.tif', dpi=600, format='tif')

# DO NOT SHOW PLOT BEFORE SAVING IT
plt.show() # not needed when working in notebooks
```

### 3D plots

```python
import numpy as np
import matplotlib.pyplot as plt
import numpy.random as R
from scipy.integrate import odeint

# Lorenz attractor
def lorenz(x, t):
    sig  = 10
    beta = 8/3
    rho  = 28

    dxdt = np.zeros(3)
    dxdt[0] = sig*(x[1]-x[0])
    dxdt[1] = x[0]*(rho-x[2])-x[1]
    dxdt[2] = x[0]*x[1] - beta*x[2]

    return dxdt

x0 = np.array([1., 1., 1.])

dt = .01; tmax = 40
tims = np.arange(0,tmax,dt)
x = odeint(lorenz, x0, tims)
```

```python
fig = plt.figure(); ax = plt.axes() 
ax.plot(x[:,0],x[:,1])
ax.set_xlabel('x'); ax.set_ylabel('y')
ax.set_title('Lorenz Attractor');

plt.show()
```

```python
fig = plt.figure(); ax = plt.axes() 
ax.plot(x[:,0],x[:,2])
ax.set_xlabel('x'); ax.set_ylabel('z')
ax.set_title('Lorenz Attractor');

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot(x[:,0], x[:,1], x[:,2]);

plt.show()
```

### multivariate normal

```python
def my_mvnpdf(x, M, S):
    return (1 / np.sqrt((2 * np.pi * np.linalg.det(S)))) * np.exp(-(1/2) * (np.transpose(x-M) @ np.linalg.inv(S) @ (x-M)))

mx = 0; my = 0
sx = 1; sy = 2; r = .4
mu = np.array([mx, my])
sig = np.array([[sx**2,   r*sx*sy],
                [r*sx*sy, sy**2]])

xy = np.array([3, 2])
                                                         
print(my_mvnpdf(xy, mu, sig))
```

```python
# make 3D plot
# https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html

inc = .25
lim = 7
xsteps = np.arange(-lim, lim, inc)
ysteps = np.arange(-lim, lim, inc)
X, Y = np.meshgrid(xsteps, ysteps)
Z = np.zeros(X.shape)

for i in range(len(xsteps)):
    for j in range(len(ysteps)):
        xy = np.array([X[i,j], Y[i,j]])
        Z[i,j] = my_mvnpdf(xy, mu, sig)
```

```python
# surface plot
# https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z)

ax.set_xlim((-lim, lim)); ax.set_xticks(np.arange(-lim, lim+1, 2))
ax.set_ylim((-lim, lim)); ax.set_yticks(np.arange(-lim, lim+1, 2));

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='autumn')

ax.set_xlim((-lim, lim)); ax.set_xticks(np.arange(-lim, lim+1, 2))
ax.set_ylim((-lim, lim)); ax.set_yticks(np.arange(-lim, lim+1, 2));

plt.show()
```

```python
# contour plot
# https://jakevdp.github.io/PythonDataScienceHandbook/04.04-density-and-contour-plots.html

fig = plt.figure()
ax = plt.axes()

ax.contour(X, Y, Z)
ax.axis('square')

ax.set_xlim((-lim, lim)); ax.set_xticks(np.arange(-lim, lim+1, 2))
ax.set_ylim((-lim, lim)); ax.set_yticks(np.arange(-lim, lim+1, 2));

plt.show()
```

```python
fig = plt.figure()
ax = plt.axes()

ax.contourf(X, Y, Z)
ax.axis('square')

ax.set_xlim((-lim, lim)); ax.set_xticks(np.arange(-lim, lim+1, 2))
ax.set_ylim((-lim, lim)); ax.set_yticks(np.arange(-lim, lim+1, 2));

plt.show()
```

```python
mx = 0; my = 0
sx = 2; sy = 1; r = -.7
M  = np.array([mx, my])
S  = np.array([[sx**2,   r*sx*sy],
               [r*sx*sy, sy**2]])
data = R.multivariate_normal(M, S, size=2000)

fig, ax = plt.subplots()

xm = 10
ax.scatter(data[:,0], data[:,1])
ax.axis('square')
ax.set_xlim((-xm, xm)); ax.set_xticks(np.arange(-xm, xm+1, 2))
ax.set_ylim((-xm, xm)); ax.set_yticks(np.arange(-xm, xm+1, 2));

plt.show()
```
