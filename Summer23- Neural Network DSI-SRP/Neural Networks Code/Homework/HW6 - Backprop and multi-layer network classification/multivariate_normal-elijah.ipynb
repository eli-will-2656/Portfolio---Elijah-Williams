# Multivariate Normal Distribution
*Normal distribution generalized to higher dimensions*

Looking at the components of each N-vector, they form a single-variable normal distribution

N-d samples can be drawn from a multivariate normal distribution using `numpy`

```python
random.multivariate_normal(mean, cov, size=None, check_valid='warn', tol=1e-8)
```

* Mean = Most likely value that values will be drawn around
* cov = Square Matrix that describes how $x_1,x_2,...,x_n$ variables correlate with eachother (slope of line of best fit between $x_i,x_j$
   * Must be `symmetric` (equal to its transpose) and `positive-semidefinite`

````python
import numpy as np
from numpy import random

mean = [0, 0]
cov = [[1, 0], [0, 1]]  # diagonal covariance

import matplotlib.pyplot as plt
x, y = np.random.multivariate_normal(mean, cov, 500).T
plt.plot(x, y, 'x')
plt.axis('equal')
plt.show()
plt.hist(x)
plt.title("X-axis")
plt.show()
plt.hist(y)
plt.title("y-axis")
plt.show()
````

We can change the covariance to generate different shaped normal distributions
For example, if we want there to be .5 correlation between x and y

````python
cov = np.array(
    [[1, 0],
    [0, 1]]
)

x, y = np.random.multivariate_normal(mean, cov, 500).T

plt.plot(x,y,'x') # Here we use x to specify points
plt.axis('equal')
plt.show()

````
